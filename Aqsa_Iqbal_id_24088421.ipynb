```python

# Convert columns to appropriate data types
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')
df['in_deezer_playlists'] = pd.to_numeric(df['in_deezer_playlists'], errors='coerce')
df['in_shazam_charts'] = pd.to_numeric(df['in_shazam_charts'], errors='coerce')

df = df.dropna()

# Scatter plot between Streams and Spotify_playlists
plt.figure(figsize=(12, 7))
sns.scatterplot(x='streams', y='in_spotify_playlists', data=df)
plt.title('Scatter Plot of Streams vs Spotify Playlists')
plt.xlabel('Streams')
plt.ylabel('Spotify Playlists')
plt.show()

# Bar plot of Key
plt.figure(figsize=(10, 6))
sns.countplot(x='key', data=df)
plt.title('Bar Plot of Key')
plt.xlabel('Key')
plt.ylabel('Count')
plt.show()

numeric_columns = df.select_dtypes(include=[np.number])

# Correlation heatmap
plt.figure(figsize=(12, 8))
corr = numeric_columns.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Calculate statistical moments for streams
mean_streams = df['streams'].mean()
variance_streams = df['streams'].var()
skewness_streams = df['streams'].skew()
kurtosis_streams = df['streams'].kurtosis()

print(f"Mean: {mean_streams}")
print(f"Variance: {variance_streams}")
print(f"Skewness: {skewness_streams}")
print(f"Kurtosis: {kurtosis_streams}")

# Histogram of 'streams'
plt.figure(figsize=(16, 9))
sns.histplot(df['streams'], kde=True, bins=30)
plt.title('Distribution of Streams')
plt.xlabel('Streams')
plt.ylabel('Frequency')
plt.show()

# Box plot of 'streams'
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['streams'])
plt.title('Box Plot of Streams')
plt.xlabel('Streams')
plt.show()

# Discussion
print("\nDiscussion of Statistical Moments:")
print(f"1. Mean: The average number of streams is {mean_streams:.2f}, indicating that most songs have around this number of streams.")
print(f"2. Variance: The variance is {variance_streams:.2f}, suggesting that the number of streams varies across songs.")
print(f"3. Skewness: The skewness is {skewness_streams:.2f}, indicating a positively skewed distribution with a few songs having extremely high streams.")
print(f"4. Kurtosis: The kurtosis is {kurtosis_streams:.2f}, which means the distribution has heavy tails and outliers (songs with exceptionally high or low streams).")
```
